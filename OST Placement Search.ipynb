```python
# Import required libraries
import PyPDF2
# import textract
import re
import string
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Python program showing a use of input()
  
val = input("Enter your folder path: ")
```

```python
import os
```

```python
os.chdir(val)
```

```python
file_names = os.listdir(val)
```

```python
space = "//"
file_names_2 = []
for i in range(0,len(file_names)):
    file = val + space + file_names[i]
    file_names_2.append(file)
```

```python
text_files = []
```

```python
for k in range(0,len(file_names_2)):
    # Open pdf file
    pdfFileObj = open(file_names_2[k],'rb')
    # Read file
    pdfReader = PyPDF2.PdfFileReader(pdfFileObj)
    # Get total number of pages
    num_pages = pdfReader.numPages
    # Initialize a count for the number of pages
    count = 0
    # Initialize a text empty etring variable
    text = ""
    # Extract text from every page on the file
    while count < num_pages:
        pageObj = pdfReader.getPage(count)
        count +=1
        text += pageObj.extractText()
    # Convert all strings to lowercase
    text = text.lower()
    # Remove numbers
    text = re.sub(r'\d+','',text)
    # Remove punctuation
    text = text.translate(str.maketrans('','',string.punctuation))
    text_files.append(text)
```

```python
text_files_splitted = []
for x in range(0,len(text_files)):
    text = text_files[x].split()
    text_files_splitted.append(text)
```

```python
candidate_names = []
for z in range(0,len(file_names_2)):
    candidate = text_files_splitted[z][0] + " " + text_files_splitted[z][1]
    candidate_names.append(candidate)
```

```python
candidate_details = pd.DataFrame({"Candidate_names":candidate_names,"File Name":file_names})
```
